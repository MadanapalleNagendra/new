## `4. Pre-Processing:`  Our collected data is cleaned this stage.

```python
df.drop(columns=["Title_Based_Price","PG_INFO"],inplace=True)
```

  5: *Identify outliers in numeric columns*:
#### Outlier detection using IQR method for numeric price columns
    **Display the cleaned data sample and identified outliers**
    **Number of outliers per column**

    An outlier is a value that is much higher or lower than most of the other values in a dataset. It's like a number that stands out because it      doesn't fit in with the rest, often being far away from the average.

```python
numeric_cols = [ 'Price']
outliers = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers[col] = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))][col]


data_cleaned_sample = df.head()
outliers_summary = {col: len(outliers[col]) for col in outliers}  
```

```python
print("outliers_summary:",outliers_summary)
```

1. Distribution of Price (Histogram)
To visualize the distribution of the Price column (assuming it's numerical), a histogram is suitable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Plot the distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, color='teal')  # `kde=True` adds a kernel density estimate line
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## `Word Cloud:` 
A Word Cloud (also called Tag Cloud) is a visual representation of text data where the size of each word indicates its frequency or importance. Words that appear more frequently are displayed larger, while words that appear less frequently are smaller. Word clouds are commonly used for displaying the most frequent terms in a dataset, such as in text analysis, feedback summaries, or for visualizing keywords from a collection of documents.

`How to Use Word Cloud in Python`
You can use the WordCloud library in Python to generate word clouds from categorical data like the Title, Living, Location, City, or Share_type columns.

```python
from wordcloud import WordCloud
```

```python
# Create a Word Cloud for the 'City' column
text = ' '.join(df['City'])

# Create a word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

# Display the Word Cloud using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')  # Interpolation to smooth the image
plt.axis('off')  # Turn off axis
plt.title('Word Cloud of Cities')
plt.show()
```

#### 1. Price vs. Sharing Type: Compare prices for AC and non-AC accommodations.
You can use a box plot to compare the distribution of prices for different sharing types (e.g., AC and non-AC).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Create a Box Plot for Price vs. Sharing Type
plt.figure(figsize=(20, 6))
sns.boxplot(x='Share_type', y='Price', data=df, palette='coolwarm')  # Use palette for colors
plt.title('Price vs. Sharing Type (AC vs Non-AC)')
plt.xlabel('Sharing Type')
plt.ylabel('Price')
plt.show()
```

#### Violin Plot (Alternative to Box Plot):
* A violin plot gives a more detailed view of the distribution of Price for different sharing types.


```python
# Create a Violin Plot for Price vs. Sharing Type
plt.figure(figsize=(20, 6))
sns.violinplot(x='Share_type', y='Price', data=df, palette='muted')
plt.title('Price Distribution by Sharing Type')
plt.xlabel('Sharing Type')
plt.ylabel('Price')
plt.show()
```

#### 2. City vs. Price: Identify cities with the most affordable or expensive accommodations.
* To analyze Price vs. City, you can use a  bar plot. This will allow you to see the price distribution across different cities.
* The box plot shows the distribution of prices across different cities. The width of the box shows the interquartile range, and the line inside the box represents the median price.

```python
# Create a Bar Plot for Average Price by City
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Price', data=df, palette='crest')  # `crest` is a color palette
plt.title('Average Price per City')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

## `1 General Business Strategy:`

```python
demand_by_city = df.groupby(['City', 'Living']).size().reset_index(name='Count')

# Visualize demand
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Count', hue='Living', data=demand_by_city)
plt.title('Demand for Accommodation Types by City')
plt.xlabel('City')
plt.ylabel('Number of Accommodations')
plt.legend(title='Living Type')
plt.xticks(rotation=45)
plt.show()
```

```python
pricing_analysis = df.groupby(['Living', 'Location']).agg({
    'Price': ['mean', 'min', 'max']
}).reset_index()

# Rename columns for clarity
pricing_analysis.columns = ['Living', 'Location', 'Avg_Price', 'Min_Price', 'Max_Price']

sns.kdeplot(data=pricing_analysis, x="Avg_Price", hue="Living", fill=True, common_norm=False, alpha=0.5)

print(pricing_analysis)
```

```python

plt.figure(figsize=(14, 7))
sns.boxplot(x='Living', y='Avg_Price', data=pricing_analysis)
plt.title('Price Distribution by Living Type')
plt.xlabel('Living Type')
plt.ylabel('Price')
plt.show()




```
